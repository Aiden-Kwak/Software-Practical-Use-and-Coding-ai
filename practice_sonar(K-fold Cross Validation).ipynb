```python
from keras.models import Sequential, load_model
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder

import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
seed = 0
np.random.seed(seed)
tf.random.set_seed(3)

df = pd.read_csv('D:\\ai\\Software Practical Use and Coding\\13\\sonar.csv', header=None)
```

```python
df
```

```python
dataset = df.values
X = dataset[:,0:60]
Y_obj = dataset[:,60]

e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

K겹 교차검증->sklearn에서 제공하는 StratifiedKFold 함수를 사용

```python
from sklearn.model_selection import StratifiedKFold

n_fold = 5
skf = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed) 
```

정확도(Accuracy)를 매번 저장하여 한 번에 보여줄 수 있도록 리스트를 사용하여 저장

```python
# 정확도(Accuracy)를 매번 저장하여 한 번에 보여줄 수 있게 accuracy 리스트를 생성
accuracy = []
step=0

# 아래의 for문을 n_fold만큼 반복되게 함
# skf.split(X, Y)는 데이터셋을 학습 데이터셋과 테스트 데이터셋으로 나눈 후, 각각 train과 test에 할당
# n_fold 마다 셔플이 적용되어 학습 데이터셋과 테스트 데이터셋이 교차될 수 있다. 

for train, test in skf.split(X, Y):  # train, test로 분리됨 
    step += 1
    print("● K :  %d 번째 iteration " %step)
    
    # 모델 설정
    # 모델을 만들고 실행하는 부분을 위의 for 구문내에서 n_fold만큼 반복되게 함
    model = Sequential()
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    
    # 모델 컴파일    
    #model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
    model.compile(loss='binary_crossentropy',  optimizer='adam', metrics=['accuracy'])
    
    # 모델 실행
    # 학습 데이터셋을 기반으로 학습
    model.fit(X[train], Y[train], epochs=10, batch_size=5, verbose=1)    
    
    # 테스트 데이터셋을 기반으로 검증 
    k_loss, k_accuracy = model.evaluate(x=X[test], y=Y[test], verbose=1)        # 테스트 데이터셋 loss, accuracy 
    print("Evaluate with test dataset : Loss = {:.4f}".format(k_loss))          # 테스트 데이터셋의 loss 출력
    print("Evaluate with test dataset : Accuracy = {:.4f}".format(k_accuracy))  # 테스트 데이터셋의 accuracy 출력
    
    # accuracy 리스트에 정확도(Accuracy)를 저장
    accuracy.append(k_accuracy)
```

정확도(Accuracy)를 K에 해당하는 횟수만큼 출력

```python
print("K-fold Cross Validation을 사용한 딥러닝")
print("K 값 :  %.f " % n_fold)
print("Accuracy : ", accuracy) # accuracy 리스트 값 출력 
print("Accuracy의 K 번 iteration 마다 정확도를 의미한다.")
```
