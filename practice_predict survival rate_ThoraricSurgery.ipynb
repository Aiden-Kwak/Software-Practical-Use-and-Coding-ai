폐암 수술 환자 생존율 예측. 데이터 출처: 폴란드 브로츠와프 의과대학 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
import numpy as np
import tensorflow as tf
```

```python
seed=0
np.random.seed(3)
tf.random.set_seed(3)
```

```python
print(np.random.seed(3))
print(tf.random.set_seed(3))
```

```python
Data_set=np.loadtxt("D:\\ai\\Software Practical Use and Coding\\12\\ThoraricSurgery.csv", delimiter=",")
```

```python
Data_set
```

```python
X = Data_set[:,0:17]
Y = Data_set[:,17]
```

```python
# 학습 데이터 70 %, 테스트 데이터셋 30% 로 설정
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=seed)
```

```python
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, Y_train, epochs=100, batch_size=10) # 학습 데이터셋 329 개로 학습
```

```python
# 30%에 해당하는 141개의 테스트 데이터셋으로 정확도 계산 결과를 출력

print("\n Test Accuracy : %.4f" % (model.evaluate(X_test, Y_test)[1]))  #테스트 데이터셋 141 개로 정확도 계산 
```

딥러닝 학습모델을 이용한 테스트. <br>
predict- 테스트 데이터를 이용한 성능평가-> 학습에 활용안한 테스트 데이터셋 쓰는게<br> 바람직할것

```python
import numpy as np
kim = np.array([[293,1,3.8,2.8,0,0,0,0,0,0,12,0,0,0,1,0,62]])
print(model.predict(kim))

test_kim  = model.predict(kim)*100
print("백분율로 표시 : ",test_kim) 
print("Kim 폐암 수술 후 생존율 예측 : %.2f" %test_kim, "%")
```
