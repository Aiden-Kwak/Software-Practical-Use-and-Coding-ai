```python
from keras.models import Sequential, load_model
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder

import pandas as pd
import numpy
import tensorflow as tf
```

```python
seed = 0
numpy.random.seed(seed)
tf.random.set_seed(3)

df = pd.read_csv('D:\\ai\\Software Practical Use and Coding\\13\\sonar.csv', header=None)
```

```python
dataset = df.values
X = dataset[:,0:60]  # 60개의 속성
Y_obj = dataset[:,60]  # 마지막 열에 해당하는 1개는 클래스
```

```python
print(type(Y_obj))
Y_obj  # 마지막열에 해당하는 1개의 클래스는 숫자형이 아님
```

텍스트 형식의 클래스를 숫자로 변환

```python
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
```

```python
Y
```

Holdout Cross Validation 이용

```python
from sklearn.model_selection import train_test_split

# 홀드 아웃 교차 검증(Holdout Cross Validation)
# 총 208개의 데이터셋 중에서 145개의 학습 데이터셋, 검증을 위해 63개의 테스트 데이터셋 준비

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=seed)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
model = Sequential()
model.add(Dense(24,  input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',  optimizer='adam',    metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=100, verbose=2)
```

```python
a,b = model.evaluate(x=X_test, y=Y_test, verbose=0) # 63개의 테스트 데이터셋 결과 
print("Evaluate with test dataset : Loss = {:.4f}".format(a))      # 테스트 데이터셋의 loss 
print("Evaluate with test dataset : Accuracy = {:.4f}".format(b))  # 테스트 데이터셋의 accuracy
```
