```python
import codecs 
# 학습시키기 위해서 positive, negative, neutral로 클래스를 만들어보려했는데 
#데이터를 하나하나 하려니 너무 오래걸려서 
#positive단어를 담은 파일과 negative단어를 담은 파일을 각각 만들어 자동화시켜본다
#codecs는 만든 txt파일을 만지기위해 임포트하였다
```

```python
positive=[]
negative=[]
binding_list=[]

pos=codecs.open("D:\\ai\\기말대체\\positive_stock.txt",'rb',encoding='UTF-8')
neg=codecs.open("D:\\ai\\기말대체\\negative_stock.txt",'rb',encoding='UTF-8')
```

```python
while True:
    line=pos.readline()
    line=line.replace('\r\n','')
    positive.append(line)
    binding_list.append(line)
    if not line: break
pos.close()
del positive[len(positive)-1]
del binding_list[len(binding_list)-1]
```

```python
while True:
    line=neg.readline()
    line=line.replace('\r\n','')
    negative.append(line)
    binding_list.append(line)
    if not line: break
neg.close()
del negative[len(negative)-1]
del binding_list[len(binding_list)-1]
```

```python
import csv
import glob

path="D:\\ai\\기말대체2\\csv_stock\\"
merge_path="D:\\ai\\기말대체2\\merge_stock.csv"

file_list=glob.glob(path+'*')
with open(merge_path,'w',encoding='UTF-8') as f:
    for file in file_list:
        with open(file,'r',encoding='UTF-8') as f2:
            while True:
                line=f2.readline()
                if not line:
                    break
                f.write(line)
        file_name=file.split('\\')[-1]
        print(file.split('\\')[-1]+' write complete!')
print('****file merge complete')
```
